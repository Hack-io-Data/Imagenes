 Los conjuntos y su intersección están claramente identificados con los colores y las etiquetas adecuadas.

 Azul A={2,3}

 Amarillo B={2,5}
 
 Verde A∩B={2}

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Definir los conjuntos A y B
A = {2, 3}
B = {2, 5}

# Crear el diagrama de Venn
plt.figure(figsize=(8, 8))
venn = venn2([A, B], ('A', 'B'))

# Colorear las áreas
venn.get_label_by_id('10').set_text('A = {2, 3}')
venn.get_label_by_id('01').set_text('B = {2, 5}')
venn.get_label_by_id('11').set_text('A ∩ B = {2}')

# Establecer colores
venn.get_patch_by_id('10').set_color('blue')
venn.get_patch_by_id('01').set_color('yellow')
venn.get_patch_by_id('11').set_color('green')

# Establecer transparencia
venn.get_patch_by_id('10').set_alpha(0.6)
venn.get_patch_by_id('01').set_alpha(0.6)
venn.get_patch_by_id('11').set_alpha(0.6)

# Título del diagrama
plt.title('Intersección de los Conjuntos A y B')

# Mostrar el diagrama
plt.show()
```

```python
# Crear el diagrama de Venn para la Unión de los Conjuntos A y B
plt.figure(figsize=(8, 8))
venn_union = venn2([A, B], ('A', 'B'))

# Título del diagrama
plt.title('Unión de los Conjuntos A y B')

# Colorear las áreas
venn_union.get_label_by_id('10').set_text('A = {2, 3}')
venn_union.get_label_by_id('01').set_text('B = {2, 5}')
venn_union.get_label_by_id('11').set_text('A ∪ B = {2, 3, 5}')

# Establecer colores
venn_union.get_patch_by_id('10').set_color('green')
venn_union.get_patch_by_id('01').set_color('green')
venn_union.get_patch_by_id('11').set_color('green')

# Establecer transparencia
venn_union.get_patch_by_id('10').set_alpha(0.6)
venn_union.get_patch_by_id('01').set_alpha(0.6)
venn_union.get_patch_by_id('11').set_alpha(0.6)

# Mostrar el diagrama
plt.show()
```

Intersección (Verde): Esta sección está vacía porque solo nos enfocamos en las diferencias específicas.

```python
# Crear el diagrama de Venn para la Diferencia de los Conjuntos A y B (A - B)
plt.figure(figsize=(8, 8))
venn_difference = venn2([A, B], ('A', 'B'))

# Colorear las áreas
venn_difference.get_label_by_id('10').set_text('A - B = {3}')
venn_difference.get_label_by_id('01').set_text('B - A = {5}')
venn_difference.get_label_by_id('11').set_text('')

# Establecer colores
venn_difference.get_patch_by_id('10').set_color('blue')
venn_difference.get_patch_by_id('01').set_color('yellow')
venn_difference.get_patch_by_id('11').set_color('green')

# Establecer transparencia
venn_difference.get_patch_by_id('10').set_alpha(0.6)
venn_difference.get_patch_by_id('01').set_alpha(0.6)
venn_difference.get_patch_by_id('11').set_alpha(0.6)

# Título del diagrama
plt.title('Diferencia de los Conjuntos A y B')

# Mostrar el diagrama
plt.show()
```

La intersección (donde A y B se superponen) está invisible, resaltando solo los elementos únicos en cada conjunto.

```python
# Crear el diagrama de Venn para la Diferencia Simétrica de los Conjuntos A y B (A Δ B)
plt.figure(figsize=(8, 8))
venn_symmetric_difference = venn2([A, B], ('A', 'B'))

# Colorear las áreas
venn_symmetric_difference.get_label_by_id('10').set_text('A - B = {3}')
venn_symmetric_difference.get_label_by_id('01').set_text('B - A = {5}')
venn_symmetric_difference.get_label_by_id('11').set_text('')

# Establecer colores
venn_symmetric_difference.get_patch_by_id('10').set_color('blue')
venn_symmetric_difference.get_patch_by_id('01').set_color('yellow')
venn_symmetric_difference.get_patch_by_id('11').set_color('white')  # La intersección será invisible

# Establecer transparencia
venn_symmetric_difference.get_patch_by_id('10').set_alpha(0.6)
venn_symmetric_difference.get_patch_by_id('01').set_alpha(0.6)
venn_symmetric_difference.get_patch_by_id('11').set_alpha(0.0)  # Hacer la intersección invisible

# Título del diagrama
plt.title('Diferencia Simétrica de los Conjuntos A y B')

# Mostrar el diagrama
plt.show()
```
